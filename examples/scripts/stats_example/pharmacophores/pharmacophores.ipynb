```python
import os
import warnings
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
import scienceplots
import scipy.stats as stats
import pingouin as pg
from scipy.stats import binomtest, wilcoxon
from matplotlib.ticker import MaxNLocator

try:
    plt.style.use(["science", "no-latex", "nature"])
except Exception:
    plt.style.use("default")

# ── CONFIGURATION ────────────────────────────────────────────────────────────
DIRECTORY        = "sim_results"
DESCRIPTORS_FILE = os.path.join(DIRECTORY, "cyp3a4_only_ligs.csv")

TARGET_FILES = [
    "rfaa_plip_filtered.csv",
    "gnina_new_plip.csv",
    "gdock_new_plip.csv",
    "vina_new_plip.csv",
]

INTERACTION_TYPES = [
    "Hydrophobic", "HBond", "WaterBridge", "SaltBridge",
    "PiStack", "PiCation", "Halogen", "Metal",
]

DESCRIPTORS_TO_PLOT = [
    ("MolWt", "MW"),
    ("NumRotatableBonds", "RotBonds"),
]

EXPECTED_ROWS = 235
START_FRAME   = 25.0
END_FRAME     = 259.0
FRAME_TO_NS   = 20 / 1000
TAU_FIG_SIZE  = (2.1, 2.5)
TAU_DPI       = 600

# ── DATA LOADER ──────────────────────────────────────────────────────────────
def get_program_from_filename(filename):
    base = os.path.basename(filename)
    clean = base.replace(".csv", "").replace("_new_plip", "").replace("_plip_filtered", "")
    return clean.strip().upper()

def load_and_validate_data(directory, file_list):
    all_blocks = []
    print(f"Loading {len(file_list)} files from '{directory}'...")
    for filename in file_list:
        filepath = os.path.join(directory, filename)
        if not os.path.exists(filepath):
            print(f"  [Warning] Not found: {filepath}")
            continue
        program_name = get_program_from_filename(filename)
        try:
            df = pd.read_csv(filepath, skipinitialspace=True)
            df.columns = df.columns.str.strip()
            if "System_Name" not in df.columns:
                continue
            try:
                df["Ligand_ID"] = df["System_Name"].astype(str).str.split("-").str[1].str.strip()
            except Exception:
                df["Ligand_ID"] = "UNKNOWN"
            df["File_Group_ID"] = (df["System_Name"] != df["System_Name"].shift()).cumsum()
            for _, group_df in df.groupby("File_Group_ID"):
                total_rows = len(group_df)
                for i in range(0, total_rows, EXPECTED_ROWS):
                    chunk = group_df.iloc[i: i + EXPECTED_ROWS].copy()
                    if len(chunk) != EXPECTED_ROWS:
                        continue
                    if "Frame" in chunk.columns:
                        chunk = chunk.sort_values("Frame").reset_index(drop=True)
                        min_f, max_f = chunk["Frame"].min(), chunk["Frame"].max()
                        if not (np.isclose(min_f, START_FRAME) and np.isclose(max_f, END_FRAME)):
                            continue
                    chunk["Program"] = program_name
                    sys_name = chunk["System_Name"].iloc[0]
                    chunk["Unique_Block_ID"] = f"{program_name}_{sys_name}_{i}"
                    mid_idx = EXPECTED_ROWS // 2
                    chunk["Simulation_Part"] = np.nan
                    chunk["Simulation_Part"] = chunk["Simulation_Part"].astype(object)
                    chunk.loc[0: mid_idx - 1, "Simulation_Part"] = "1st Half"
                    chunk.loc[mid_idx + 1:,   "Simulation_Part"] = "2nd Half"
                    chunk = chunk.dropna(subset=["Simulation_Part"])
                    all_blocks.append(chunk)
        except Exception as e:
            print(f"  [Error] {filename}: {e}")
    if not all_blocks:
        return pd.DataFrame()
    return pd.concat(all_blocks, ignore_index=True)

df_full = load_and_validate_data(DIRECTORY, TARGET_FILES)
if not df_full.empty:
    print(f"Loaded {len(df_full)} rows across {df_full['Unique_Block_ID'].nunique()} simulation blocks.")
else:
    print("No valid data loaded.")

# ── STATISTICS ───────────────────────────────────────────────────────────────

def run_sign_test(y_start, y_end):
    """Paired sign test (binomial) testing for DECREASE. Returns p_two, p_loss, prop_decreased, bootstrap CI string."""
    diffs = np.array(y_end) - np.array(y_start)
    n_loss  = np.sum(diffs < 0)  # Count instances where value decreased
    n_valid = np.sum(diffs != 0)
    
    if n_valid == 0:
        return 1.0, 1.0, 0.5, "[N/A]"
        
    p_two  = binomtest(n_loss, n_valid, p=0.5, alternative="two-sided").pvalue
    # Alternative 'greater' here tests if the proportion of losses is significantly > 0.5
    p_loss = binomtest(n_loss, n_valid, p=0.5, alternative="less").pvalue 
    prop_decr = n_loss / n_valid

    # Bootstrap 95% CI on proportion decreased
    ci_str = "[N/A]"
    if len(diffs) >= 2 and hasattr(stats, "bootstrap"):
        def stat_prop_decr(x, axis=-1):
            n_l = np.sum(x < 0, axis=axis)
            n_v = np.sum(x != 0, axis=axis)
            with np.errstate(divide="ignore", invalid="ignore"):
                out = n_l / n_v
            if isinstance(out, np.ndarray):
                out[n_v == 0] = 0.5
            return out
        try:
            with warnings.catch_warnings():
                warnings.simplefilter("ignore")
                res = stats.bootstrap((diffs,), stat_prop_decr, confidence_level=0.95,
                                      n_resamples=5000, random_state=42, method="BCa")
                if not np.isnan(res.confidence_interval.low):
                    ci_str = "[{:.2f}, {:.2f}]".format(
                        res.confidence_interval.low, res.confidence_interval.high)
        except Exception:
            pass
    return p_two, p_loss, prop_decr, ci_str


def run_wilcoxon_within(y_start, y_end):
    """Wilcoxon signed-rank test on paired differences. Returns p-value."""
    diffs = np.array(y_end) - np.array(y_start)
    if len(diffs) < 10 or np.all(diffs == 0):
        return np.nan
    try:
        _, p = wilcoxon(diffs, alternative="two-sided", zero_method="wilcox")
        return p
    except Exception:
        return np.nan


def run_mwu_between(va, vb):
    """Mann-Whitney U via pingouin. Returns p_two, p_less, CLES."""
    if len(va) < 3 or len(vb) < 3:
        return np.nan, np.nan, np.nan
    try:
        res_two  = pg.mwu(va, vb, alternative="two-sided")
        res_less = pg.mwu(va, vb, alternative="less")
        return (res_two["p-val"].values[0],
                res_less["p-val"].values[0],
                res_less["CLES"].values[0])
    except Exception:
        return np.nan, np.nan, np.nan


def _direction(prop_decr):
    if prop_decr > 0.55: return "LOSS"
    if prop_decr < 0.45: return "GAIN"
    return "Stable"


def print_stats_table(df_wide, value_col, label):
    """
    Global / interaction-type stats table.
    Within:  Sign test only (p_two, p_loss, % decreased + bootstrap CI).
    """
    unique_cats = sorted(df_wide["_category"].unique())
    sep = "=" * 120
    print("\n" + sep)
    print(("STATISTICS: " + label).center(120))
    print(sep)

    print("\n[Within-Program: Sign Test (1st Half -> 2nd Half)]")
    hdr = "{:<20} | {:<5} | {:<12} | {:<12} | {:<20} | {:<7} | Result".format(
        "Program", "N", "p(2-sided)", "p(greater)", "% Decr [95% CI]", "Sig?")
    print(hdr)
    print("-" * len(hdr))

    for cat in unique_cats:
        cat_data = df_wide[df_wide["_category"] == cat]
        y_start  = cat_data[1].values
        y_end    = cat_data[2].values
        if len(y_start) == 0:
            continue
        p_two, p_loss, prop_decr, ci_str = run_sign_test(y_start, y_end)
        sig = "YES" if p_two < 0.05 else "NO"
        print("{:<20} | {:<5} | {:<12.2e} | {:<12.2e} | {:.3f} {:<14} | {:<7} | {}".format(
            cat, len(y_start), p_two, p_loss,
            prop_decr, ci_str, sig, _direction(prop_decr)))


def print_descriptor_stats_table(df_wide, category_col, value_col, label,
                                 low_lbl, high_lbl):
    """
    Descriptor (MW / RotBonds) stats table.
    Within each group:  Sign test + Wilcoxon (1st -> 2nd Half).
    Between Low vs High: Mann-Whitney U via pg.mwu (p_two, p_less, CLES).
    """
    sep = "=" * 130
    print("\n" + sep)
    print(("DESCRIPTOR STATISTICS: " + label).center(130))
    print(sep)

    print("\n[Within-Group: Sign Test + Wilcoxon Signed-Rank (1st Half -> 2nd Half)]")
    hdr = "{:<28} | {:<5} | {:<12} | {:<12} | {:<20} | {:<12} | Result".format(
        "Group", "N", "p(2-sided)", "p(loss)", "% Decr [95% CI]", "Wilcoxon p")
    print(hdr)
    print("-" * len(hdr))

    group_means = {}
    for lbl in [low_lbl, high_lbl]:
        cat_data = df_wide[df_wide["_category"] == lbl]
        y_start  = cat_data[1].values
        y_end    = cat_data[2].values
        if len(y_start) == 0:
            continue
        p_two, p_loss, prop_decr, ci_str = run_sign_test(y_start, y_end)
        p_wilcox = run_wilcoxon_within(y_start, y_end)
        w_str = "{:.2e}".format(p_wilcox) if not np.isnan(p_wilcox) else "N/A"
        sig   = "YES" if p_two < 0.05 else "NO"
        print("{:<28} | {:<5} | {:<12.2e} | {:<12.2e} | {:.3f} {:<14} | {:<12} | {} ({})".format(
            lbl[:28], len(y_start), p_two, p_loss,
            prop_decr, ci_str, w_str, _direction(prop_decr), sig))
        group_means[lbl] = (y_start + y_end) / 2

    # Between Low vs High
    if low_lbl in group_means and high_lbl in group_means:
        p2, pl, cles = run_mwu_between(group_means[low_lbl], group_means[high_lbl])
        p2_s   = "{:.2e}".format(p2)   if not np.isnan(p2)   else "N/A"
        pl_s   = "{:.2e}".format(pl)   if not np.isnan(pl)   else "N/A"
        cles_s = "{:.3f}".format(cles) if not np.isnan(cles) else "N/A"
        sig    = "YES" if (not np.isnan(p2) and p2 < 0.05) else "NO"
        print("\n[Between-Group: Mann-Whitney U — Low vs High (pg.mwu)]")
        print("  p(2-sided): {}  |  p(less): {}  |  CLES: {}  |  Significant: {}".format(
            p2_s, pl_s, cles_s, sig))

# ── PAIRED BOX-SLOPE ─────────────────────────────────────────────────────────
def plot_paired_slope_style(df, category_col, value_col, title, filename_suffix, print_stats=True):
    if df.empty:
        return
    df_agg = df.groupby([category_col, "Unique_Block_ID", "Simulation_Part"])[value_col].mean().reset_index()
    df_agg["Simulation_Part"] = df_agg["Simulation_Part"].map({"1st Half": 1, "2nd Half": 2})
    df_agg = df_agg.dropna(subset=["Simulation_Part"])
    df_agg["Simulation_Part"] = df_agg["Simulation_Part"].astype(int)
    df_wide = df_agg.pivot(
        index=[category_col, "Unique_Block_ID"],
        columns="Simulation_Part", values=value_col
    ).dropna().reset_index()
    df_wide["_category"] = df_wide[category_col]
    
    if print_stats:
        print_stats_table(df_wide, value_col, title)
        
    unique_cats = sorted(df_wide[category_col].unique())
    n_cats = len(unique_cats)
    if n_cats == 0:
        return
    muted = sns.color_palette("muted")
    c_start, c_end = muted[0], muted[1]
    fig, axes = plt.subplots(1, n_cats, figsize=(4 * n_cats, 5))
    if n_cats == 1:
        axes = [axes]
        
    for i, cat in enumerate(unique_cats):
        ax = axes[i]
        cat_data = df_wide[df_wide[category_col] == cat]
        if cat_data.empty:
            continue
        y_start  = cat_data[1].values
        y_end    = cat_data[2].values
        n_points = len(y_start)
        pos_box1, pos_pts1, pos_pts2, pos_box2 = -0.4, 0.0, 1.0, 1.4
        jitter = np.random.uniform(-0.06, 0.06, n_points)
        x_s = pos_pts1 + jitter
        x_e = pos_pts2 + jitter
        for j in range(n_points):
            ax.plot([x_s[j], x_e[j]], [y_start[j], y_end[j]],
                    color="gray", alpha=0.1, linewidth=0.8, zorder=1)
        ax.scatter(x_s, y_start, s=15, color=c_start, alpha=0.6, zorder=2)
        ax.scatter(x_e, y_end,   s=15, color=c_end,   alpha=0.6, zorder=2)
        bp1 = ax.boxplot([y_start], positions=[pos_box1], widths=0.25, patch_artist=True, showfliers=False, zorder=3)
        bp2 = ax.boxplot([y_end],   positions=[pos_box2], widths=0.25, patch_artist=True, showfliers=False, zorder=3)
        for patch in bp1["boxes"]:
            patch.set_facecolor(c_start); patch.set_alpha(0.5)
        for patch in bp2["boxes"]:
            patch.set_facecolor(c_end); patch.set_alpha(0.5)
        for bp in [bp1, bp2]:
            for elem in ["whiskers", "caps", "medians"]:
                plt.setp(bp[elem], color="black", linewidth=1.5)
                
        ax.set_xlabel(cat + "\n(N=" + str(n_points) + ")", fontsize=16)
        ax.set_xticks([pos_box1, pos_pts1, pos_pts2, pos_box2])
        ax.set_xticklabels(["", "", "", ""])
        ax.set_ylim(-0.05, 1.05)
        if i == 0:
            ax.set_ylabel("Fraction Retained", fontsize=16)
            ax.tick_params(axis="y", labelsize=14)
        else:
            ax.set_ylabel("")
            ax.set_yticklabels([])
        ax.tick_params(axis="both", which="both", bottom=False, left=False, top=False, right=False)
        for sp in ["top", "right", "left"]:
            ax.spines[sp].set_visible(False)
        ax.spines["bottom"].set_visible(True)
        ax.yaxis.grid(False)
        
    plt.suptitle(title + " Evolution", fontsize=18, y=1.02)
    plt.tight_layout()
    plt.savefig(filename_suffix + ".png", dpi=300, bbox_inches="tight")
    print("Saved: " + filename_suffix + ".png")
    plt.show()
    plt.close()

# ── RUN BOX-SLOPE PLOTS ───────────────────────────────────────────────────────
if not df_full.empty and "Fraction_Retained" in df_full.columns:
    plot_paired_slope_style(df_full, "Program", "Fraction_Retained",
                            "Global Interaction Retention", "global_retention_slope")

for int_type in INTERACTION_TYPES:
    col_cons = int_type + "_Conserved"
    col_lost = int_type + "_Lost"
    if col_cons not in df_full.columns:
        continue
    subset = df_full.copy()
    total  = subset[col_cons] + subset.get(col_lost, 0)
    subset["Calculated_Fraction"] = np.where(total > 0, subset[col_cons] / total, np.nan)
    subset = subset.dropna(subset=["Calculated_Fraction"])
    if not subset.empty:
        plot_paired_slope_style(subset, "Program", "Calculated_Fraction",
                                int_type + " Retention", "retention_slope_" + int_type)

if os.path.exists(DESCRIPTORS_FILE):
    df_desc   = pd.read_csv(DESCRIPTORS_FILE)
    right_key = "PDB ID"
    if right_key in df_desc.columns:
        df_desc[right_key]    = df_desc[right_key].astype(str).str.strip().str.upper()
        df_full["Ligand_ID"]  = df_full["Ligand_ID"].astype(str).str.strip().str.upper()
        merged = pd.merge(df_full, df_desc, left_on="Ligand_ID", right_on=right_key, how="inner")
        if not merged.empty:
            for col, label in DESCRIPTORS_TO_PLOT:
                if col not in merged.columns:
                    continue
                median_val = merged[col].median()
                low_lbl    = "Low "  + label + " (<= " + "{:.2f}".format(median_val) + ")"
                high_lbl   = "High " + label + " (> "  + "{:.2f}".format(median_val) + ")"
                merged["Desc_Group"] = merged[col].apply(
                    lambda x, lo=low_lbl, hi=high_lbl: lo if x <= median_val else hi
                )
                
                df_agg_desc = merged.groupby(
                    ["Desc_Group", "Unique_Block_ID", "Simulation_Part"]
                )["Fraction_Retained"].mean().reset_index()
                df_agg_desc["Simulation_Part"] = df_agg_desc["Simulation_Part"].map(
                    {"1st Half": 1, "2nd Half": 2})
                df_agg_desc = df_agg_desc.dropna(subset=["Simulation_Part"])
                df_agg_desc["Simulation_Part"] = df_agg_desc["Simulation_Part"].astype(int)
                df_wide_desc = df_agg_desc.pivot(
                    index=["Desc_Group", "Unique_Block_ID"],
                    columns="Simulation_Part", values="Fraction_Retained"
                ).dropna().reset_index()
                df_wide_desc["_category"] = df_wide_desc["Desc_Group"]
                
                print_descriptor_stats_table(
                    df_wide_desc, "Desc_Group", "Fraction_Retained",
                    label, low_lbl, high_lbl)
                    
                plot_paired_slope_style(merged, "Desc_Group", "Fraction_Retained",
                                        "Retention by " + label, "descriptor_slope_" + label,
                                        print_stats=False)

# ── VIOLIN PLOTS ─────────────────────────────────────────────────────────────
def _violin_base(df, x_col, y_col, ylabel, filename, order=None):
    program_order = order or sorted(df[x_col].unique())
    plt.figure(figsize=(10, 6))
    sns.violinplot(
        data=df, x=x_col, y=y_col, hue="Simulation_Part",
        split=True, hue_order=["1st Half", "2nd Half"],
        order=program_order, palette="muted",
        inner="quartile", linewidth=1.0, cut=0
    )
    plt.xlabel("" if x_col == "Desc_Group" else x_col, fontsize=16)
    plt.ylabel(ylabel, fontsize=16)
    plt.tick_params(axis="both", which="major", labelsize=14)
    plt.ylim(-0.05, 1.05)
    if plt.gca().legend_:
        plt.gca().legend_.remove()
    plt.tight_layout()
    plt.savefig(filename, dpi=300)
    print("Saved: " + filename)
    plt.show()
    plt.close()

if not df_full.empty and "Fraction_Retained" in df_full.columns:
    _violin_base(df_full, "Program", "Fraction_Retained",
                 "Fraction Retained", "global_retention_violins.png")

for int_type in INTERACTION_TYPES:
    col_cons = int_type + "_Conserved"
    col_lost = int_type + "_Lost"
    if col_cons not in df_full.columns:
        continue
    subset = df_full.copy()
    total  = subset[col_cons] + subset.get(col_lost, 0)
    subset["Calculated_Fraction"] = np.where(total > 0, subset[col_cons] / total, np.nan)
    subset = subset.dropna(subset=["Calculated_Fraction"])
    if not subset.empty:
        _violin_base(subset, "Program", "Calculated_Fraction",
                     "Fraction Retained", "interaction_retention_" + int_type + "_violins.png")

if os.path.exists(DESCRIPTORS_FILE):
    df_desc   = pd.read_csv(DESCRIPTORS_FILE)
    right_key = "PDB ID"
    if right_key in df_desc.columns:
        df_desc[right_key]   = df_desc[right_key].astype(str).str.strip().str.upper()
        df_full["Ligand_ID"] = df_full["Ligand_ID"].astype(str).str.strip().str.upper()
        merged = pd.merge(df_full, df_desc, left_on="Ligand_ID", right_on=right_key, how="inner")
        if not merged.empty:
            for col, label in DESCRIPTORS_TO_PLOT:
                if col not in merged.columns:
                    continue
                median_val = merged[col].median()
                low_lbl    = "Low "  + label + " (<= " + "{:.2f}".format(median_val) + ")"
                high_lbl   = "High " + label + " (> "  + "{:.2f}".format(median_val) + ")"
                merged["Desc_Group"] = merged[col].apply(
                    lambda x, lo=low_lbl, hi=high_lbl: lo if x <= median_val else hi
                )
                _violin_base(merged, "Desc_Group", "Fraction_Retained",
                             "Global Fraction Retained",
                             "descriptor_retention_" + label + "_violins.png",
                             order=[low_lbl, high_lbl])

# ── TAU HISTOGRAM + ACF DECAY ─────────────────────────────────────────────────
if not df_full.empty:
    unique_progs   = sorted(df_full["Program"].unique())
    program_colors = dict(zip(unique_progs, sns.color_palette("colorblind", n_colors=len(unique_progs))))
    mk_results     = []
    all_acf_curves = []

    for (prog, block_id, sys_name), group in df_full.groupby(["Program", "Unique_Block_ID", "System_Name"]):
        if len(group) < 20:
            continue
        group = group.sort_values("Frame")
        y = group["Fraction_Retained"].values
        with warnings.catch_warnings():
            warnings.simplefilter("ignore")
            tau, _ = stats.kendalltau(np.arange(len(y)), y)
        mk_results.append({"Program": prog, "Tau": tau if not np.isnan(tau) else 0.0})
        y_det = y - np.mean(y)
        if np.std(y_det) == 0:
            continue
        full_corr = np.correlate(y_det, y_det, mode="full")
        full_corr = full_corr[full_corr.size // 2:]
        acf_vals  = full_corr / full_corr[0]
        for i, val in enumerate(acf_vals[:len(y) // 2]):
            all_acf_curves.append({"Program": prog, "Time_Lag_ns": i * FRAME_TO_NS, "ACF": val})

    df_mk     = pd.DataFrame(mk_results)
    df_curves = pd.DataFrame(all_acf_curves)

    # Tau histogram
    fig1, ax1 = plt.subplots(figsize=TAU_FIG_SIZE, dpi=TAU_DPI)
    sns.histplot(data=df_mk, x="Tau", hue="Program", palette=program_colors,
                 element="step", kde=False, legend=False, ax=ax1)
    for sp in ax1.spines.values():
        sp.set_visible(True); sp.set_linewidth(0.8)
    t_min, t_max = df_mk["Tau"].min(), df_mk["Tau"].max()
    x_margin = (t_max - t_min) * 0.02 if t_max != t_min else 0.1
    ax1.set_xlim(t_min - x_margin, t_max + x_margin)
    visible_ticks = [t for t in [-0.5, 0, 0.5] if t_min <= t <= t_max]
    ax1.set_xticks(visible_ticks)
    ax1.set_xticklabels(visible_ticks, fontsize=7)
    ax1.tick_params(axis="both", which="both", labelsize=7, length=0)
    ax1.axvline(0, color="black", linewidth=0.8, alpha=0.5)
    ax1.set_xlabel(r"Kendall's $\tau$", fontsize=8) 
    ax1.set_ylabel("Count", fontsize=8)
    plt.tight_layout(pad=0.2)
    plt.savefig("plot_tau_pharmacophore.png", dpi=TAU_DPI, bbox_inches="tight")
    plt.show()
    plt.close(fig1)

    # ACF decay
    if not df_curves.empty:
        fig2, ax2 = plt.subplots(figsize=TAU_FIG_SIZE, dpi=TAU_DPI)
        sns.lineplot(data=df_curves, x="Time_Lag_ns", y="ACF", hue="Program",
                     palette=program_colors, errorbar=("ci", 95),
                     legend=False, linewidth=1.0, ax=ax2)
        for sp in ax2.spines.values():
            sp.set_visible(True); sp.set_linewidth(0.8)
        ax2.tick_params(axis="both", which="both", labelsize=7, length=0)
        ax2.set_xlim(0, df_curves["Time_Lag_ns"].max())
        ax2.margins(x=0, y=0.05)
        ax2.axhline(0, color="firebrick", linestyle="--", alpha=0.7, linewidth=0.8)
        ax2.set_xlabel("Lag Time (ns)", fontsize=8)
        ax2.set_ylabel("ACF C(t)", fontsize=8)
        plt.tight_layout(pad=0.2)
        plt.savefig("plot_acf_pharmacophore.png", dpi=TAU_DPI, bbox_inches="tight")
        plt.show()
        plt.close(fig2)
```
